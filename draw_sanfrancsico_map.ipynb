```python
import osmnx as ox
```

```python
san_francisco = ox.graph_from_place('San Francisco, USA')
ox.plot_graph(san_francisco)
```

```python
san_francisco_drive = ox.graph_from_place('San Francisco, USA', network_type='drive')
ox.plot_graph(san_francisco_drive)
```

```python
london = ox.graph_from_place('London, UK', which_result=2)
ox.plot_graph(london)
```

```python
london_drive = ox.graph_from_place('London, UK', which_result=2, network_type='drive')
ox.plot_graph(london_drive)
```

```python
hamburg_drive = ox.graph_from_place('Hamburg, Germany', which_result=2, network_type='drive')
ox.plot_graph(hamburg_drive)
```
