```python
import cartopy.crs as ccrs
import cartopy.io.img_tiles as cimgt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
cabspotting = pd.read_pickle('data_cabspotting/cabspotting.pkl')
```

```python
plt.figure(figsize=(18, 18))
img = cimgt.GoogleTiles()
ax = plt.axes(projection = img.crs)
ax.add_image(img, 5, interpolation = 'bicubic')

plt.scatter(cabspotting['longitude'], cabspotting['latitude'], s=0.25, transform=ccrs.PlateCarree())
```
